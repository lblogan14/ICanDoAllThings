# FastAPI Tutorial 12 - PUT Request

In previous sections, we have explored GET and POST requests in FastAPI to retrieve and create resources, respectively. In this section, we will learn how to use PUT requests to update existing resources.

```python
@app.put("/shipment")
def shipment_update(
    id: int, content: str, weight: float, status: str
) -> dict[str, Any]:
    shipments[id] = {
        'weight': weight,
        'content': content,
        'status': status,
    }

    return shipments[id]
```

To successfully update a resource using a PUT request, we need to define all the fields of the resource in the request. This means that when updating a shipment, we must provide all the necessary fields such as `id`, `content`, `weight`, and `status`. For example, if we want to update a shipment with `id=1`, we need to send a PUT request with the complete data for that shipment.
