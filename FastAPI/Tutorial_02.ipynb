# FastAPI Tutorial 2

As our first FastAPI application, we will create a simple API that manages the shipment.

````python
from fastapi import FastAPI

app = FastAPI()

@app.get("/shipment")
def get_shipment():
    return {
        'content': 'wooden table',
        'status': 'in transit',
    }
````

`"/shipment"` endpoint will allow us to create and retrieve shipment information.

We do not need to provide an absolute path like `"http://domain.co/shipment"`, but just the relative path `"/shipment"`.

We can also have a nested path, for example, `"/seller/shipment"` so that only sellers can access the shipment information.

````python
@app.get("/seller/shipment")
def get_shipment():
    return {
        'content': 'wooden table',
        'status': 'in transit',
    }
````

To get access to the API, we can use a browser and navigate to `http://localhost:8000/shipment` after starting the FastAPI server.

To start the server, we can first navigate to the folder where it contains our `main.py` file and use the FastAPI CLI command:

```bash
fastapi dev
```
after activating the virtual environment.

We can check the logs to see if the server has started successfully and additionally, we can see the URL where the server is running.
